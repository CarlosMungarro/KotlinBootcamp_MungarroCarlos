```kotlin

fun whatShouldIDoToday(mood: String, weather: String = "sunny", temperature: Int = 24): String {
    return when {
        mood == "happy" && weather == "sunny" -> "go for a walk"
        mood == "sad" && weather == "rainy" && temperature < 15 -> "stay in bed"
        temperature > 35 -> "go swimming"
        mood == "happy" && weather == "rainy" -> "watch a movie"
        mood == "sad" -> "drink tea and relax"
        weather == "sunny" -> "have a picnic"
        else -> "stay home and read"
    }
}

// Funci√≥n main
fun main() {
    println("=== Activity Suggester ===")

    // Probar diferentes
    println(whatShouldIDoToday("happy"))
    println(whatShouldIDoToday("sad"))
    println(whatShouldIDoToday("happy", "rainy"))
    println(whatShouldIDoToday("sad", "rainy", 10))
    println(whatShouldIDoToday("neutral", "sunny", 40))
}


main()
```
