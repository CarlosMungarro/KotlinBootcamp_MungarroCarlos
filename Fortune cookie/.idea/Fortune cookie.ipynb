```kotlin

fun getFortuneForTest(birthday: Int): String {
    return when (birthday) {
        28, 31 -> "Special day fortune!"
        in 1..7 -> "First week fortune!"
        else -> "Regular fortune #${birthday % 7}"
    }
}

println("=== Different Birthdays ===")
val testDays = listOf(1, 5, 7, 15, 28, 31, 32)

for (day in testDays) {
    println("Day $day: ${getFortuneForTest(day)}")
}
```
