```kotlin
// Operación en una línea encadenada
val fishLeft = 2.plus(71).plus(233).minus(13)
println("Peces restantes: $fishLeft")

// Acuarios necesarios (30 peces por acuario)
val tanksNeeded = (fishLeft + 30 - 1) / 30
println("Acuarios necesarios: $tanksNeeded")

// Bonus: ¿Qué tienen de especial los números?
println("Bonus: Los números 2, 71, 233, 13 y 293 son primos")
```

```kotlin
// Variable mutable
var rainbowColor = "Red"
rainbowColor = "Naranja"
println("rainbowColor: $rainbowColor")

// Variable inmutable
val blackColor = "Negro"
println("blackColor: $blackColor")
// blackColor = "Azul"
```

```kotlin

// Dos formas de declarar variables nullables
var greenColor: String? = null
var blueColor: String? = null
println("greenColor: $greenColor")
println("blueColor: $blueColor")

// Otra forma explícita
var blueColor2: String?
blueColor2 = null
println("blueColor2: $blueColor2")
```

```kotlin
// 1. Lista con dos elementos null - Forma 1
val list1 = listOf(null, null)
println("list1: $list1")

// Forma 2
val list2: List<Int?> = listOf(null, null)
println("list2: $list2")

// Forma 3
val list3 = mutableListOf<String?>(null, null)
println("list3: $list3")

// 2. Lista que es null
val nullList: List<String>? = null
println("nullList: $nullList")

// O también
var anotherNullList: List<Int>? = null
println("anotherNullList: $anotherNullList")
```
