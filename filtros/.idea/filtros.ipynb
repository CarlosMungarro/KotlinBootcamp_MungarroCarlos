```kotlin
// 1. Crear lista
val spices = listOf("curry", "pepper", "cayenne", "ginger", "red")

// 2. Filtrar curries y ordenar por longitud
val curries = spices
    .filter { it.contains("curry") }
    .sortedBy { it.length }
println("Curries ordenadas: $curries")

// 3. Filtrar que empiezan con 'c' y terminan con 'e'
val cToE = spices.filter { it.startsWith('c') && it.endsWith('e') }
println("Empiezan con c y terminan con e: $cToE")

// 4. Primeros 3 que empiezan con 'c'
val firstThreeC = spices.take(3).filter { it.startsWith('c') }
println("Primeros 3 que empiezan con c: $firstThreeC")
```
