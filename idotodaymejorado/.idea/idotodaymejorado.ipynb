```kotlin

fun isHappySunny(mood: String, weather: String) = mood == "happy" && weather == "sunny"
fun isSadRainyCold(mood: String, weather: String, temp: Int) = mood == "sad" && weather == "rainy" && temp < 15
fun isVeryHot(temp: Int) = temp > 35
fun isTired(mood: String) = mood == "tired" // NUEVA situación 1
fun isSnowy(weather: String, temp: Int) = weather == "snowy" && temp < 0 // NUEVA situación 2
fun isWeekendFun(mood: String, weather: String) = mood == "excited" && weather == "sunny" // NUEVA situación 3

// FUNCIÓN PRINCIPAL MEJORADA
fun whatShouldIDoToday(mood: String, weather: String = "sunny", temperature: Int = 24): String {
    return when {
        isHappySunny(mood, weather) -> "go for a walk"
        isSadRainyCold(mood, weather, temperature) -> "stay in bed"
        isVeryHot(temperature) -> "go swimming"
        isTired(mood) -> "take a nap" // NUEVA actividad 1
        isSnowy(weather, temperature) -> "build a snowman" // NUEVA actividad 2
        isWeekendFun(mood, weather) -> "go to a party" // NUEVA actividad 3
        mood == "happy" && weather == "rainy" -> "watch a movie"
        mood == "sad" -> "drink tea and relax"
        weather == "sunny" -> "have a picnic"
        else -> "stay home and read"
    }
}

// 3. CHALLENGE: OBTENER MOOD DEL USUARIO
fun getMoodFromUser(): String {
    print("How are you feeling? (happy/sad/tired/excited): ")
    return readLine()!!.lowercase()
}

// FUNCIÓN MAIN CON TODO
fun main() {
    println("=== Activity Suggester MEJORADO ===")

    // Probar las 3 situaciones nuevas
    println("1. Tired mood: ${whatShouldIDoToday("tired")}")
    println("2. Snowy weather: ${whatShouldIDoToday("neutral", "snowy", -5)}")
    println("3. Excited & sunny: ${whatShouldIDoToday("excited", "sunny")}")

    println("\n=== Prueba con usuario ===")
    val userMood = getMoodFromUser()
    println("You should: ${whatShouldIDoToday(userMood)}")
}


main()
```
