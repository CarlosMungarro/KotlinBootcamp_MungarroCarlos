```kotlin
// Task 2 - Ejemplos con when

// Ejemplo 1: Nombre fijo
println("=== Evaluando nombres de peces ===")
val fishName = "Nemo"
println("Nombre: '$fishName'")

when (fishName.length) {
    0 -> println("Resultado: Error - nombre vacío")
    in 3..12 -> println("Resultado: Good fish name")
    else -> println("Resultado: OK fish name")
}

// Ejemplo 2: Probando varios nombres
println("\n--- Probando diferentes nombres ---")

val testNames = listOf("", "Sal", "Tilapia", "A", "Clownfish", "Hippocampus")

for (name in testNames) {
    val result = when (name.length) {
        0 -> " Error: vacío"
        in 3..12 -> " Buen nombre"
        else -> " Nombre OK"
    }
    println("'$name' (${name.length} letras) → $result")
}
```

```kotlin
// Función reusable para evaluar nombres
fun evaluarNombrePez(nombre: String): String {
    return when (nombre.length) {
        0 -> " ERROR: Nombre vacío"
        in 3..12 -> " BUEN nombre"
        else -> " Nombre aceptable"
    }
}

// Probando la función
println("=== Usando función reusable ===")
println("'Goldfish': ${evaluarNombrePez("Goldfish")}")
println("'': ${evaluarNombrePez("")}")
println("'Tiburón': ${evaluarNombrePez("Tiburón")}")
```
