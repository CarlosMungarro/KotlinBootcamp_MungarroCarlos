```kotlin
// 1. primero se crea un array de enteros del 11 al 15
val numbers = arrayOf(11, 12, 13, 14, 15)

println("Array de numeros: ${numbers.joinToString()}")

// 2. Crear una lista mutable vacia para Strings
val stringList = mutableListOf<String>()
println("Lista inicial (vacía): $stringList")

// 3. Bucle for para agregar representaciones de strings
for (number in numbers) {
    stringList.add(number.toString())
}

println("Lista despues del bucle: $stringList")
println("Elementos en la lista: ${stringList.size}")
```

```kotlin
// Otra version

val numbers2 = (11..15).toList()
println("Numeros: $numbers2")

val stringList2 = numbers2.map { it.toString() }.toMutableList()
println("Lista convertida: $stringList2")

// O usando forEach
val stringList3 = mutableListOf<String>()
numbers2.forEach { stringList3.add(it.toString()) }
println("Lista con forEach: $stringList3")
```

```kotlin
// Desafío

println("=== Numeros entre 0 y 100 divisibles por 7 ===")

// Metodo 1: Bucle for
val divisibleBySeven = mutableListOf<Int>()
for (i in 0..100) {
    if (i % 7 == 0) {
        divisibleBySeven.add(i)
    }
}
println("Con bucle for: ${divisibleBySeven.joinToString()}")

// Metodo 2: Usando range y step
val divisibleBySeven2 = (0..100 step 7).toList()
println("Con step 7: ${divisibleBySeven2.joinToString()}")

// Metodo 3: Usando filter
val divisibleBySeven3 = (0..100).filter { it % 7 == 0 }
println("Con filter: ${divisibleBySeven3.joinToString()}")


println("\nTotal de numeros encontrados: ${divisibleBySeven.size}")
println("Primer número: ${divisibleBySeven.first()}")
println("Ultimo número: ${divisibleBySeven.last()}")
```
